# Lebel Encoding

```python
from sklearn.preprocessing import LabelEncoder
lebel = LabelEncoder()#Using for level encoding
```

```python
from pandas.core.dtypes.common import is_numeric_dtype#showing all value in numeric form
```

```python

for col in df1.columns:
    if is_numeric_dtype(df1[col]):
        continue
    
    df1[col] = lebel.fit_transform(df1[col])
```

```python
df1.head()
```

# One Hot Encoding

```python
columns = ['sex']

for col in columns:
    one_hot = pd.get_dummies(df2[col],drop_first=True, prefix='new')
    new_df2 = pd.concat([df2 ,one_hot],axis=1).drop(col,axis=1)
```

```python
new_df2.head(10)
```

```python
columns = ['smoker']

for col in columns:
    one_hot = pd.get_dummies(df3[col],drop_first=True, prefix='smoker')
    new_df3 = pd.concat([new_df2 ,one_hot],axis=1).drop(col,axis=1)
```

```python
new_df3.head(10)
```

```python
columns = ['region']

for col in columns:
    one_hot = pd.get_dummies(df4[col],drop_first=True, prefix='new')
    new_df4 = pd.concat([new_df3 ,one_hot],axis=1).drop(col,axis=1)
```

```python
new_df4.head(10)#in this line we see all encoded columns
```
